```python
import graphlab as gl
import numpy
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
sales = gl.SFrame('../week1-simple-regression/kc_house_data.gl/')
```

```python
sales.sort(['sqft_living', 'price'])
```

```python
def polynomial_sframe(feature, degree):
    poly_sframe = gl.SFrame()
    poly_sframe['power_1'] = feature
    if degree > 1:
        for power in range(2, degree+1): 
            # first we'll give the column a name:
            name = 'power_' + str(power)
            # then assign poly_sframe[name] to the appropriate power of feature
            poly_sframe[name] = gl.SArray(feature).apply(lambda x: x**power)
    return poly_sframe

def polynomial_model(feature, degree, penalty):
    poly_data = polynomial_sframe(sales[feature], degree)
    my_features = poly_data.column_name()
    poly_data['price'] = sales['price']
    model = gl.linear_regression.create(poly_data, features=my_features, target='price', l2_penalty=penalty, validation_set=None,verbose=False)
    return model
```

```python
model = polynomial_model(sales['sqft_living'], 15)
```
